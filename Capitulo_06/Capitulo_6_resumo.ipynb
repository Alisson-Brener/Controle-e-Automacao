# Capítulo 6
## A Estabilidade de Sistemas Lineares com Realimentação

Este capítulo aborda a questão mais fundamental no projeto de sistemas de controle: a **estabilidade**. Um sistema de controle só é útil se for estável. Analisamos o que define um sistema estável e apresentamos um método poderoso para verificar a estabilidade sem a necessidade de calcular as raízes da equação característica.

### 1. O Conceito de Estabilidade
A estabilidade é a propriedade mais importante de um sistema dinâmico. Um sistema é considerado estável se sua resposta a uma entrada limitada for também limitada (estabilidade **BIBO - Bounded-Input, Bounded-Output**).

* **Estabilidade Absoluta:** Define se um sistema é estável ou instável. A resposta é binária: sim ou não.
* **Estabilidade Relativa:** Mede o *grau* de estabilidade. Um sistema altamente oscilatório, embora tecnicamente estável, é menos estável (ou possui menor estabilidade relativa) do que um sistema bem amortecido.

A estabilidade de um sistema linear está diretamente ligada à localização dos polos de sua função de transferência em malha fechada (ou aos autovalores da matriz de estados **A**).
* **Polos no semiplano esquerdo do plano-s:** Resultam em uma resposta que decai com o tempo. **O sistema é estável**.
* **Polos no semiplano direito do plano-s:** Resultam em uma resposta que cresce exponencialmente. **O sistema é instável**.
* **Polos simples (não repetidos) no eixo imaginário ($j\omega$):** Resultam em uma oscilação sustentada. **O sistema é marginalmente estável**.
* **Polos repetidos no eixo imaginário ($j\omega$):** Resultam em uma resposta que cresce com o tempo (ex: $t \cdot \sin(\omega t)$). **O sistema é instável**.

### 2. O Critério de Estabilidade de Routh-Hurwitz
Este é um método analítico que permite determinar a estabilidade de um sistema sem calcular explicitamente os polos. Ele responde a duas perguntas:
1.  O sistema é estável?
2.  Se não for, quantos polos estão no semiplano direito?

O critério baseia-se na análise dos coeficientes da equação característica do sistema, $q(s) = a_n s^n + a_{n-1} s^{n-1} + \dots + a_0 = 0$.

**Condição Necessária (mas não suficiente):** Para um sistema ser estável, **todos os coeficientes $a_i$ devem existir e ter o mesmo sinal** (geralmente todos positivos). Se algum coeficiente for nulo ou negativo, o sistema é instável e a análise pode parar aqui.

**Condição Necessária e Suficiente:** É a aplicação da **Tabela de Routh**. A tabela é construída a partir dos coeficientes da equação característica. O critério afirma que:
> O número de raízes com parte real positiva (polos no semiplano direito) é igual ao número de trocas de sinal na primeira coluna da Tabela de Routh.

Para que um sistema seja estável, não deve haver nenhuma troca de sinal na primeira coluna.

#### Casos Especiais na Tabela de Routh:
* **Caso 1: Zero na primeira coluna (e outros elementos não nulos na linha):** Substitui-se o zero por um número pequeno e positivo, $\epsilon$, e continua-se a construção da tabela. Analisa-se o sinal dos elementos da primeira coluna quando $\epsilon \to 0$.
* **Caso 2: Uma linha inteira de zeros:** Isso indica que existem raízes simétricas em relação à origem (ex: pares em $\pm j\omega$ ou $\pm \sigma$). Para continuar, utiliza-se o **polinômio auxiliar**, $U(s)$, formado a partir da linha *acima* da linha de zeros. A derivada desse polinômio, $\frac{dU(s)}{ds}$, é usada para substituir a linha de zeros e completar a tabela. As raízes do polinômio auxiliar são também raízes da equação característica original.

### 3. Estabilidade Relativa
O critério de Routh-Hurwitz pode ser estendido para analisar a estabilidade relativa. Para verificar se todas as raízes têm uma parte real menor que um valor $-\sigma_1$, faz-se uma **mudança de variável** $s = s_n - \sigma_1$ na equação característica. Em seguida, aplica-se o critério de Routh-Hurwitz ao novo polinômio em $s_n$. Se o novo sistema for estável, significa que todas as raízes do sistema original estão à esquerda da linha $s = -\sigma_1$.

### 4. Estabilidade de Sistemas em Variáveis de Estado
Para um sistema representado no espaço de estados por $\dot{\mathbf{x}} = \mathbf{Ax} + \mathbf{Bu}$, a estabilidade é determinada pelos autovalores da matriz **A**. A equação característica é dada por:
$$\det(s\mathbf{I} - \mathbf{A}) = 0$$
As raízes desta equação (os autovalores de **A**) são os polos do sistema. Se todos os autovalores tiverem parte real negativa, o sistema é estável.
