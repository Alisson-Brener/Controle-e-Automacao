## O Método do Lugar das Raízes (Root Locus)

O Método do Lugar das Raízes (LGR), introduzido por W.R. Evans, é uma técnica gráfica poderosa que plota a trajetória das raízes da equação característica no plano-s à medida que um parâmetro do sistema (geralmente o ganho $K$) varia de zero a infinito.

### 1. O Conceito Fundamental
O comportamento dinâmico de um sistema em malha fechada é determinado pelos seus polos. A equação característica de um sistema com realimentação é dada por:
$$1 + KG(s)H(s) = 0$$

Isso implica duas condições que qualquer ponto no Lugar das Raízes deve satisfazer:
1.  **Critério de Magnitude:** $|KG(s)H(s)| = 1$
2.  **Critério de Ângulo:** $\angle G(s)H(s) = 180^\circ \pm k360^\circ$ (para $K > 0$)

O LGR nos mostra visualmente se o sistema se torna instável (raízes cruzando para o semiplano direito) ou se torna mais/menos oscilatório (mudança no $\zeta$) ao ajustarmos $K$.

### 2. Procedimento para Esboçar o LGR
Embora computadores façam isso instantaneamente, entender as regras de construção é vital para o projeto intuitivo. Os 7 passos principais são:

1.  **Início e Fim:** O lugar começa nos polos de malha aberta ($K=0$) e termina nos zeros de malha aberta ($K=\infty$).
2.  **Eixo Real:** Um segmento do eixo real faz parte do LGR se houver um número **ímpar** de polos e zeros à sua direita.
3.  **Assíntotas:** Se o sistema tem $n$ polos e $M$ zeros, $n-M$ ramos vão para o infinito seguindo assíntotas lineares.
    * **Centroide ($\sigma_A$):** Ponto onde as assíntotas cruzam o eixo real.
        $$\sigma_A = \frac{\sum(-p_i) - \sum(-z_i)}{n - M}$$
    * **Ângulo ($\phi_A$):**
        $$\phi_A = \frac{180^\circ (2k + 1)}{n - M}$$
4.  **Cruzamento do Eixo Imaginário:** Determinado pelo critério de Routh-Hurwitz. Define o ganho crítico para estabilidade.
5.  **Pontos de Saída/Chegada (Breakaway/Break-in):** Ocorrem no eixo real onde múltiplos ramos se encontram. São os máximos ou mínimos locais da função de ganho, encontrados resolvendo $\frac{dK}{ds} = 0$.
6.  **Ângulo de Partida/Chegada:** Para polos/zeros complexos, usa-se o critério de ângulo para determinar a direção exata em que o lugar sai do polo ou chega ao zero.
7.  **Completar o Esboço:** Unir os pontos seguindo as regras de simetria.

### 3. Controladores PID e o LGR
O capítulo introduz o controlador mais usado na indústria, o PID (Proporcional-Integral-Derivativo), e como ele altera o LGR:

* **Controlador Proporcional (P):** $G_c(s) = K_P$. Apenas altera o ganho, movendo os polos ao longo do lugar original.
* **Controlador Integral (I):** Adiciona um polo na origem. Elimina o erro em regime estacionário, mas tende a desestabilizar o sistema (empurra o LGR para a direita).
* **Controlador Derivativo (D):** Adiciona um zero. Tem efeito antecipatório, adicionando amortecimento e melhorando a estabilidade (puxa o LGR para a esquerda).

A função de transferência do PID é:
$$G_c(s) = K_P + \frac{K_I}{s} + K_D s = \frac{K_D s^2 + K_P s + K_I}{s}$$
Isso introduz **um polo na origem** e **dois zeros** (que podem ser complexos ou reais), permitindo "moldar" o lugar das raízes para passar por uma região de desempenho desejado.

### 4. Sintonia de Ziegler-Nichols
São métodos empíricos para ajustar os ganhos do PID quando o modelo matemático não é preciso.
* **Método de Malha Fechada (Ganho Crítico):** Aumenta-se $K_P$ até o sistema oscilar com amplitude constante (ganho crítico $K_c$ e período $T_c$). Os ganhos finais são derivados de $K_c$ e $T_c$.
* **Método de Malha Aberta (Curva de Reação):** Baseado na resposta ao degrau da planta em malha aberta (atraso de tempo e taxa de subida).
