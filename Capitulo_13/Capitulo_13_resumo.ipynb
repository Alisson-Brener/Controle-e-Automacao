## Sistemas de Controle Digital

Este capítulo introduz o controle por computador, onde os sinais não são contínuos, mas sim amostrados em intervalos discretos de tempo $T$. Essa é a realidade da grande maioria dos sistemas de controle modernos, desde freios ABS até o piloto automático de aviões.

### 1. Sistemas com Dados Amostrados
Em um sistema de controle digital, o computador recebe dados do processo através de um **Conversor Analógico-Digital (A/D)** e envia comandos através de um **Conversor Digital-Analógico (D/A)**.
* **Amostrador (Sampler):** Converte um sinal contínuo $r(t)$ em uma sequência de pulsos $r(kT)$ nos instantes $0, T, 2T, \dots$.
* **Segurador de Ordem Zero (ZOH - Zero-Order Hold):** Mantém o valor da amostra constante até a próxima amostragem, reconstruindo um sinal "em escada" para o processo contínuo. Sua função de transferência é:
$$G_{ZOH}(s) = \frac{1 - e^{-sT}}{s}$$

### 2. A Transformada Z
Assim como a Transformada de Laplace é usada para sistemas contínuos (equações diferenciais), a Transformada Z é usada para sistemas discretos (equações de diferenças).
* **Definição:**
$$F(z) = \sum_{k=0}^{\infty} f(kT)z^{-k}$$
* **Relação com o Plano-s:** A variável $z$ está relacionada com $s$ por:
$$z = e^{sT}$$
Isso mapeia o **semiplano esquerdo** do plano-s (estabilidade contínua) para o **interior do círculo unitário** no plano-z.

### 3. Estabilidade em Sistemas Digitais
A estabilidade de um sistema discreto é determinada pelas raízes da equação característica $1 + G(z) = 0$.
* **Estável:** Todos os polos de malha fechada estão **dentro** do círculo unitário ($|z| < 1$).
* **Instável:** Pelo menos um polo está **fora** do círculo unitário ($|z| > 1$).
* **Marginalmente Estável:** Polos sobre o círculo unitário ($|z| = 1$).

### 4. Projeto de Controladores Digitais
Existem duas abordagens principais:
1.  **Projeto no Plano-s e Discretização:** Projeta-se o controlador $G_c(s)$ como se o sistema fosse contínuo e, em seguida, converte-se para $D(z)$ usando uma aproximação (como Tustin ou *Matched Pole-Zero*). É válido para frequências de amostragem altas.
2.  **Projeto Direto no Plano-z:** O modelo da planta é discretizado primeiro ($G(z) = \mathcal{Z}\{G_{ZOH}(s)G_p(s)\}$), e o controlador $D(z)$ é projetado diretamente usando o Lugar das Raízes no plano-z.

### 5. Implementação do Controlador (PID Digital)
Um controlador PID contínuo pode ser implementado digitalmente aproximando a derivada por diferenças finitas e a integral por soma de retângulos:
* **Termo Proporcional:** $u_P(k) = K_P e(k)$
* **Termo Integral:** $u_I(k) = u_I(k-1) + K_I T e(k)$
* **Termo Derivativo:** $u_D(k) = \frac{K_D}{T} [e(k) - e(k-1)]$

A saída total é a soma desses termos. Essa equação de diferenças é o que realmente roda no microcontrolador.
