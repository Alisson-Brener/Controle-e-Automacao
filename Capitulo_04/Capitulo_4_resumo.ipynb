# Capítulo 4
## Características de Sistemas de Controle com Realimentação

Este resumo aborda os conceitos essenciais do Capítulo 4 do livro "Sistemas de Controle Modernos" de Dorf & Bishop. O foco é entender as vantagens, os custos e as características fundamentais dos sistemas de controle que utilizam realimentação (malha fechada).

### 1. Introdução e Vantagens da Realimentação
Um sistema de controle em malha fechada utiliza o sinal de erro, a diferença entre a resposta desejada e a resposta real, para ajustar o processo e atingir o objetivo. Apesar de aumentar a complexidade e o custo, o controle com realimentação oferece vantagens cruciais em comparação com os sistemas em malha aberta:

* **Menor sensibilidade do sistema** a variações nos parâmetros do processo.
* **Melhor rejeição a perturbações** externas.
* **Melhor atenuação do ruído de medida** introduzido por sensores.
* **Melhor redução do erro em regime estacionário**.
* **Fácil controle e ajuste da resposta transitória** do sistema.

### 2. Análise do Sinal de Erro
O **erro de rastreamento** $E(s) = R(s) - Y(s)$ é o sinal central na análise de sistemas com realimentação, onde $R(s)$ é a entrada de referência e $Y(s)$ é a saída. A análise desse sinal revela um compromisso fundamental no projeto de controle.

Para um sistema com realimentação unitária, o erro pode ser expresso em termos do **ganho em malha aberta** $L(s) = G_c(s)G(s)$:

$$ E(s) = \frac{1}{1+L(s)}R(s) - \frac{G(s)}{1+L(s)}T_p(s) + \frac{L(s)}{1+L(s)}N(s) $$

onde $T_p(s)$ é a perturbação e $N(s)$ é o ruído de medida.

Para simplificar a análise, definimos duas funções importantes:
* **Função de Sensibilidade:** $S(s) = \frac{1}{1+L(s)}$
* **Função de Sensibilidade Complementar:** $C(s) = \frac{L(s)}{1+L(s)}$

Essas funções estão interligadas pela restrição fundamental:
$$ S(s) + C(s) = 1 $$

Isso implica que não é possível fazer com que ambas as funções sejam pequenas simultaneamente, forçando uma **solução de compromisso** no projeto:
* Para **rejeitar perturbações**, precisamos de $|L(j\omega)|$ grande, o que torna $|S(j\omega)|$ pequeno.
* Para **atenuar o ruído de medida**, precisamos de $|L(j\omega)|$ pequeno, o que torna $|C(j\omega)|$ pequeno.

### 3. Sensibilidade à Variação de Parâmetros
A sensibilidade do sistema é a razão entre a variação percentual da função de transferência do sistema, $T(s)$, e a variação percentual de um parâmetro do processo, $G(s)$.

* Em malha aberta, a sensibilidade é **1**.
* Em malha fechada, a sensibilidade é reduzida por um fator de $(1 + L(s))$:
    $$ S_G^T = \frac{1}{1+L(s)} $$

Isso mostra que, ao aumentar o ganho em malha aberta $L(s)$, o sistema se torna menos sensível a variações e incertezas nos parâmetros da planta.

### 4. Rejeição a Sinais de Perturbação
A realimentação é eficaz em reduzir o efeito de perturbações indesejadas na saída do sistema. O efeito de uma perturbação $T_p(s)$ na saída é reduzido quando o ganho em malha aberta, $|L(j\omega)|$, é grande nas frequências em que a perturbação atua (geralmente baixas frequências).

### 5. Controle da Resposta Transitória
A **resposta transitória** é a resposta do sistema em função do tempo antes de atingir o regime estacionário. Em um sistema em malha fechada, a resposta transitória (como a velocidade de resposta) pode ser facilmente ajustada alterando-se o ganho do controlador, $K_a$. Em contraste, em malha aberta, seria necessário modificar componentes físicos do processo, como o motor.

### 6. Erro em Regime Estacionário
O **erro em regime estacionário** é o erro que permanece após o decaimento da resposta transitória. Um sistema em malha fechada com um ganho estático ($L(0) = G_c(0)G(0)$) elevado terá um erro em regime estacionário significativamente menor do que um sistema em malha aberta. A realimentação monitora continuamente o erro e atua para reduzi-lo, mesmo na presença de variações de parâmetros ou erros de calibração.

### 7. Custo da Realimentação
Apesar das vantagens, a implementação da realimentação traz custos e desafios:
* **Aumento da Complexidade e Componentes:** Exige um sensor para medir a saída, que pode ser caro e introduzir ruído no sistema.
* **Perda de Ganho:** O ganho do sistema em malha fechada é reduzido por um fator de $1/(1+L(s))$ em comparação ao sistema em malha aberta.
* **Potencial de Instabilidade:** Um sistema que é estável em malha aberta pode se tornar instável quando a malha de realimentação é fechada.
